Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python

# Definir los arrays
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia
diferencia = a - b

# Calcular la media aritmética
media = np.mean(diferencia)

print("La diferencia (a - b) es:", diferencia)
print("La media aritmética de la diferencia es:", media)
```

```python

def leer_csv_nucleos_poblacion(nombre_archivo='NucleosPoblacion.csv'):
    """
    Lee un archivo CSV que contiene los núcleos de población y retorna un DataFrame.

    Parámetros:
    -----------
    nombre_archivo : str
        Nombre del archivo CSV ubicado en la misma carpeta que el script.

    Retorna:
    --------
    df : pandas.DataFrame
        DataFrame con los datos del archivo CSV.
    """
    try:
        df = pd.read_csv(nombre_archivo)
        print("Archivo leído exitosamente.")
        return df
    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo}' no se encontró en la carpeta del script.")
    except Exception as e:
        print(f"Ocurrió un error al leer el archivo: {e}")

# Ejemplo de uso:
# df_nucleos = leer_csv_nucleos_poblacion()
# print(df_nucleos.head())
```

```python
nucleos_poblacion_df = leer_csv_nucleos_poblacion('NucleosPoblacion.csv')
```

Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
# Filtrar municipios que pertenecen a la provincia de Granada
municipios_granada = nucleos_poblacion_df[nucleos_poblacion_df['Provincia'].str.lower() == 'granada']

# Extraer solo la columna de población
poblaciones = municipios_granada['Poblacion']

# Calcular las estadísticas
media = poblaciones.mean()
mediana = poblaciones.median()
desviacion = poblaciones.std()
valor_maximo = poblaciones.max()
valor_minimo = poblaciones.min()

# Mostrar los resultados
print(f"✅ Estadísticas de población en la provincia de Granada:")
print(f"- Media: {media:.2f}")
print(f"- Mediana: {mediana:.2f}")
print(f"- Desviación estándar: {desviacion:.2f}")
print(f"- Valor máximo: {valor_maximo}")
print(f"- Valor mínimo: {valor_minimo}")
```

Realice un histograma con la población de los

Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python

# ----------- OBTENER COORDENADAS DE MADRID -----------
# Buscar la fila donde el municipio es "Madrid"
madrid = nucleos_poblacion_df[nucleos_poblacion_df['Municipio'].str.lower() == 'madrid']

if madrid.empty:
    print("❌ Madrid no está en el dataset, verifica los datos.")
else:
    X_madrid = madrid.iloc[0]['X']
    Y_madrid = madrid.iloc[0]['Y']

    # ----------- CALCULAR DISTANCIA DE TODOS LOS MUNICIPIOS A MADRID -----------
    # Filtrar el dataset excluyendo Madrid
    df_sin_madrid = nucleos_poblacion_df[nucleos_poblacion_df['Municipio'].str.lower() != 'madrid'].copy()

    # Calcular la distancia euclidiana
    df_sin_madrid['Distancia'] = np.sqrt((df_sin_madrid['X'] - X_madrid)**2 + (df_sin_madrid['Y'] - Y_madrid)**2)

    # ----------- ENCONTRAR MUNICIPIOS MÁS CERCANO Y MÁS LEJANO -----------
    municipio_mas_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia'].idxmin()]
    municipio_mas_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia'].idxmax()]

    # ----------- IMPRIMIR RESULTADOS -----------
    print(f"✅ Municipio más cercano a Madrid: {municipio_mas_cercano['Municipio']} ({municipio_mas_cercano['Distancia']:.2f} unidades)")
    print(f"✅ Municipio más lejano a Madrid: {municipio_mas_lejano['Municipio']} ({municipio_mas_lejano['Distancia']:.2f} unidades)")
```
